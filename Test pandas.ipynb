## Save file with pandas

```python
import pandas as pd
import os
```

```python
df = pd.DataFrame()
df["Owoce"] = ["Jabłko", "Gruszka", "Malina", "Brzoskwinia"]
df["Warzywa"] = ["Papryka", "Pomidor", "Seler", "Pietruszka"]

# Check there is a file, if not then create
if not os.path.isfile("times.csv"):
    f = open("owoce.csv", "a")
    f.close()
    
df.to_csv("owoce.csv")
```
